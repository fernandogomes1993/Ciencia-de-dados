```python
import pandas as pd
import matplotlib.pyplot as plt

def Inflaçao():
  # base de dados
  dados = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Visualzacao_de_dados_curso_tecnico/inflaçao_acumulada_ano.xlsm')
  # removendo as duas primeiras linhas inuteis
  dados.drop([0,1],inplace=True)

  # renomeado as colunas que vou usar
  dados.rename(columns={dados.columns[0]:'Data',dados.columns[2]:'Variacao'},inplace=True)

  # memovendo o texto e deixando apenas data nas linhas
  dados['Data'] = dados['Data'].str[:4]
  # removendo linha inutil
  dados.drop(33,inplace=True)
  # transformando a coluna data em tipo data
  dados['Data'] = pd.to_datetime(dados['Data'])

  # deixando a data apenas em anos
  dados['Data'] = dados['Data'].dt.strftime('%Y')
  # pegando apenas as colunas que vou usar
  dados = dados[['Data','Variacao']]
  ultimos_10_anos_dados = dados.tail(10)


  plt.figure(figsize=(10, 6))
  plt.title('Inflaçao acumulada ultimos 10 anos %')
  plt.xticks(rotation=90)

  plt.bar(ultimos_10_anos_dados['Data'],ultimos_10_anos_dados['Variacao'],color='orange',ec="black",lw=2)

  # informaçoes sobre as barras
  # for i in range(0,10):
  #   plt.annotate(ultimos_10_anos_dados.Variacao.iloc[i],(i,ultimos_10_anos_dados.Variacao.iloc[i]+0.1),ha='center')

  # informaçoes sobre as barras colocando o simbolo %
  for i in range(0,10):
    plt.text(i,ultimos_10_anos_dados.Variacao.iloc[i]+0.1,f'{ultimos_10_anos_dados.Variacao.iloc[i]}%',ha='center')

  plt.show()

Inflaçao()







```

```python
import pandas as pd
import matplotlib.pyplot as plt2

base_selic = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Visualzacao_de_dados_curso_tecnico/selic_acumulada.xlsx')
base_selic.drop(['Reunião','Unnamed: 2','Período de vigência','Taxa Selic','Unnamed: 7'],axis=1,inplace=True)
base_selic.drop([0,1,2,3],axis=0,inplace=True)

base_selic.rename(columns={base_selic.columns[0]:'Data'},inplace=True)
base_selic.rename(columns={base_selic.columns[1]:'Meta'},inplace=True)

base_selic['Data'] = pd.to_datetime(base_selic['Data'])
base_selic.drop(columns=base_selic.columns[2],inplace=True)
base_selic = base_selic[base_selic['Data'].dt.month == 12]
base_selic['Data'] = base_selic['Data'].dt.strftime('%Y')
base_selic.head(10)

#     grafico ..................
plt2.figure(figsize=(10, 6))
plt2.xticks(rotation=90)
plt2.title('Taxa selic ultimos 10 ano %')
plt2.bar(base_selic['Data'],base_selic['Meta'],color='orange',ec="black",lw=2)
for i in range(0,10):
  plt2.text(i,base_selic.Meta.iloc[i]+0.2,f'{base_selic.Meta.iloc[i]}%',ha='center',color='blue')




plt2.show()
```
